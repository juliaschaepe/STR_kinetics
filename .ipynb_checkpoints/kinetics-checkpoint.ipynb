## Kinetics estimations and analysis

```python
#1 is C_F (cell milieu, free)
#2 is A_F (antenna, free)
#3 is SB (specifically bound)
#4 is Deg (non-specifically bound/degenerate binding)

# helpful numbers: 100 nM motif affinity, 1 uM repetitive flanks, 10-100 uM random
# start with a factor of 10 between repetitive and random flanks

factors = {}
factors['repetitive'] = 0.1
factors['random'] = 0.01
ss = {}
nuc_vol = 3e-15*1000  # cm^3
ant_vol = 3.3e-17*1000  # cm^3

for key, factor in factors.items():
    
    print('type: ', key)
    print('affinity factor: ', factor)
    
    N = 100 # number "degenerate sites"
    M = 1 # number of "motifs"
    Na = 6.022e23
    D = 10e-7  # cm^2/s
    R = np.cbrt(3*ant_vol/(4*np.pi))  # in units of cm^3
    DNA = 5e-5 # assuming 150 bp fragments in a 3 um^3 volume yeast nucleus

    Kd_in_out = 1e4 # double check this - connor had 3, slides had 4
    Kd_23 = 1e-7 # source: MITOMI data
    Kd_24 = 1e-7/(factor*N)
    Kd_34 = Kd_24/Kd_23
    Kd_12 = (1+DNA/Kd_23+DNA/Kd_24)*Kd_in_out # this shouldn't be one over

    k12 = 4*np.pi*D*R/nuc_vol # in units of cm^3/s / cm^3 = 1/s
    k21 = k12*Kd_12 # to maintain "detailed balance"
    print('k12: ', k12)
    print('k21: ', k21)

    # these are bimolecular so look a lot faster 1/Ms
    k32 = 0.015 # source: MITOMI data
    k23 = k32/Kd_23 # detailed balance
    print('k32: ', k32)
    print('k23: ', k23)

    k42 = 0.02 # source: MITOMI data - this is koff
    k24 = k42/Kd_24 # detailed balance - this is kon
    print('k42: ', k42)
    print('k24: ', k24)
    
    # these are single molecular 1/s
    k43 = 1000/(7*(N/2)) # each bp is a degenerate site ish - could play around w this
    k34 = k43/Kd_34 # detailed balance
    print('k43: ', k43)
    print('k34: ', k34)
    # this doesn't feel intuitive to me
    
    print('Kd_12: ', Kd_12) # this is 'antenna affinity'
    print('Kd_23: ', Kd_23) # this is motif affinity
    print('Kd_24: ', Kd_24) # this is "non-specific" repetitive affinity
    print('Kd_34: ', Kd_34) # this is motif affinity on 1D sliding

    # These numbers imply that Deg --> SB is rate limiting over AF --> Deg

    ss_mat = np.array([
        [-k12, k21, 0, 0],
        [0, -k23, k32, 0],
        [0, -k24, 0, k42],
        [0, 0, -k34, k43]
    ])

    ss_vec = scipy.linalg.null_space(ss_mat)/np.sum(scipy.linalg.null_space(ss_mat))
    print('equilibrium vector: ', np.round(ss_vec, 2))
    ss[key] = ss_vec
    SB = ss_vec[2][0]
    
    Deg = ss_vec[3][0]
    AF = ss_vec[1][0]
    SB_Deg = SB + Deg

    vol = ant_vol 
    MOTIF = M/(Na*vol)
    DEG = N/(Na*vol)
    TF = 1/(Na*vol) # assuming 100 molecules in a 3 um^3 volume yeast nucleus, only 1 in antenna
    TF_DNA_DEG = TF

    
    d24dt = k24*TF*DEG
    d43dt = k43*TF_DNA_DEG
    d23dt = k23*TF*MOTIF

    print('d24dt = ', d24dt)
    print('d43dt = ', d43dt)
    print('d23dt = ', d23dt)

#     k_off = (k32*SB + k42*Deg + k34*k42*SB*Deg + k43*k32*Deg*SB)/SB_Deg
#     k_off = (k32*SB + k42*Deg)
#     print('koff: ', k_off)
#     k_on = (k23*AF + k24*AF)
#     print('kon: ', k_on)
```

The conclusion for what is rate limiting is very dependent on the concentration of TF, DNA, and TF-DNA-DEG. With the volume as nuc_vol it is getting to DNA, whereas with ant_vol it is sliding. But either way, it is a bit faster to get to the motif.

With nuc_vol:

In the system of one motif, 100 flanking degenerate sites and 1 TF in antenna, getting onto DNA is rate limiting over 1D diffusion along DNA molecule. This fits with our kinetic model of binding. This model also maintains constant off rates from DNA from the flanks and from the motif regardless of repetitive or repeat. However, it does likely change the flux of off through the flanks or motif because there is a lower energy barrier to go from the motif to the repetitive flanks in comparison to random flanks.

## Gillespie simulation

```python
def get_k_array(n_deg_sites, factor, DNA=5e-5, core_affinity=1e-7, core_koff=0.015, flank_koff=0.02, velocity_prob=1000/7, Kd_in_out=1e4, tf_diffusion = 10e-7 / 100):
    nuc_vol = 3e-15
    ant_vol = 3.3e-17
    D = tf_diffusion  # cm^2/s --> L/s
    R = np.cbrt(3*ant_vol/(4*np.pi))

    Kd_23 = core_affinity # source: MITOMI data
    Kd_24 = core_affinity/(factor*n_deg_sites)
    Kd_34 = Kd_24/Kd_23
    Kd_12 = ((1+DNA/Kd_23+DNA/Kd_24)*Kd_in_out)
    
    k12 = 4*np.pi*D*R/nuc_vol
    k21 = k12*Kd_12 # to maintain "detailed balance"

    k32 = core_koff # source: MITOMI data - this is koff
    k23 = k32/Kd_23 # detailed balance - this is kon

    k42 = flank_koff # source: MITOMI data - this is koff
    k24 = k42/Kd_24 # detailed balance - this is kon

    k43 = velocity_prob/(n_deg_sites/2) # this might need to be re-evaluated
    k34 = k43/Kd_34 # detailed balance
    
    return (k12, k21, k23, k24, k32, k34, k42, k43)
```

```python
def simulate_TFsearch(T, y0, k_array, DNA=5e-5, frac_DNA=1/10):
    t = 0
    k12, k21, k23, k24, k32, k34, k42, k43 = k_array
    n_tot = np.sum(y0)
    i = 0
    y = y0
    n_rows = 100000
    sim_data = np.zeros((n_rows, len(y0)+2))
    w_mapping = [([0],[1]),
             ([1],[0]), ([1,4],[2]), ([1,5],[3]),
             ([2],[1,4]), ([2,5],[3,4]),
             ([3],[1,5]), ([3,4],[2,5])]
    
    while t < T:

        # calculates likelihood of each reaction
        w12 = k12 * y[0] * DNA
        w21 = k21 * y[1] * DNA
        # this assumes the concentration of DNA in the antenna is 1/100 of all DNA
        w23 = k23 * y[1] * y[4] * DNA
        w32 = k32 * y[2]        
        w24 = k24 * y[1] * y[5] * DNA
        w42 = k42 * y[3]        
        w34 = k34 * y[2]
        w43 = k43 * y[3]   
        # this ensures conservation of 1 motif and 100 deg sites without messing with rates
        if y[4] == 0:
            w43 = 0
        if y[5] == 0:
            w34 = 0

        # calculates tau
        w_arr = [w12, w21, w23, w24, w32, w34, w42, w43]
        W = np.sum(w_arr)
        tau = -np.log(np.random.rand()) / W
        t = t + tau

        # chooses next reaction
        rand = np.random.rand()
        for j in range(len(w_arr)):
            if rand <= (np.sum(w_arr[:j+1]) / W):
                idx_from, idx_to = w_mapping[j]
                for idx in idx_from:
                    y[idx] -= 1
                for idx in idx_to:
                    y[idx] += 1
                break
                
        # allocates more space so that sim_data is not stored dynamically
        if i >= n_rows:
            sim_data = np.vstack((sim_data, np.zeros((n_rows, len(y0)+2))))
            n_rows += n_rows
        
        # updates sim_data
        sim_data[i] = np.hstack([tau,t,y])
        i += 1
        
    np.array(sim_data)
    return sim_data[:np.argmax(sim_data[:,1])+1, :]
```

```python
# computes the fraction of time that the target is occupied
def compute_fraction_time_occupied(simulation_data, target_idx, dt_index, time_index, target_idx_2 = None):
    target_data = simulation_data[:-1,target_idx]
    if target_idx_2 is not None:
        target_data = np.add(target_data, simulation_data[:-1, target_idx_2])
    tot_occupied_time = np.sum(simulation_data[(np.where(target_data > 0)[0] + 1),dt_index])
    return tot_occupied_time/simulation_data[-1,time_index]

# computes the average occupancy of the target
def compute_mean_occupancy(simulation_data, target_idx, dt_index, target_idx_2 = None):
    if target_idx_2 is not None:
        return np.average(np.add(simulation_data[:-1,target_idx], simulation_data[:-1,target_idx_2]), weights = simulation_data[1:, dt_index])
    return np.average(simulation_data[:-1,target_idx], weights = simulation_data[1:, dt_index])
```

# Sensitivity analysis
(roughly in order of importance)
- Number of degenerate sites
- Core site affinity
- Core site kinetics (on/off rates)
- Flanking sequence kinetics (on/off rates)
- number of TF molecules
- sliding kinetics (incl. probability of binding given that a TF passes core site)
- diffusion kinetics (getting in and out of antenna volume)
- antenna volume
- DNA pseudo-concentration


```python
def simulate_MFPT(y0, k_array, DNA=5e-5):
    t = 0
    k12, k21, k23, k24, k32, k34, k42, k43 = k_array
    n_tot = np.sum(y0)
    i = 0
    y = y0
    n_rows = 100000
    sim_data = np.zeros((n_rows, len(y0)+2))
    w_mapping = [([0],[1]),
             ([1],[0]), ([1,4],[2]), ([1,5],[3]),
             ([2],[1,4]), ([2,5],[3,4]),
             ([3],[1,5]), ([3,4],[2,5])]
    
    while sim_data[:,4]:

        # calculates likelihood of each reaction
        w12 = k12 * y[0] * DNA
        w21 = k21 * y[1] * DNA
        w23 = k23 * y[1] * y[4] * DNA
        w32 = k32 * y[2]        
        w24 = k24 * y[1] * y[5] * DNA
        w42 = k42 * y[3]        
        w34 = k34 * y[2]
        w43 = k43 * y[3]   
        # this ensures conservation of 1 motif and 100 deg sites without messing with rates
        if y[4] == 0:
            w43 = 0
        if y[5] == 0:
            w34 = 0

        # calculates tau
        w_arr = [w12, w21, w23, w24, w32, w34, w42, w43]
        W = np.sum(w_arr)
        tau = -np.log(np.random.rand()) / W
        t = t + tau

        # chooses next reaction
        rand = np.random.rand()
        for j in range(len(w_arr)):
            if rand <= (np.sum(w_arr[:j+1]) / W):
                idx_from, idx_to = w_mapping[j]
                for idx in idx_from:
                    y[idx] -= 1
                for idx in idx_to:
                    y[idx] += 1
                break
                
        # allocates more space so that sim_data is not stored dynamically
        if i >= n_rows:
            sim_data = np.vstack((sim_data, np.zeros((n_rows, len(y0)+2))))
            n_rows += n_rows
        
        # updates sim_data
        sim_data[i] = np.hstack([tau,t,y])
        i += 1
        
    np.array(sim_data)
    return sim_data[:np.argmax(sim_data[:,1])+1, :]
```

### Number of degenerate sites

```python
factor = np.geomspace(1e-3,1, num=15)
N_deg = np.array([1,10,100,500,1000])

n_rep = 1
first_passage_deg = np.zeros((n_rep, len(factor), len(N_deg)))
mean_occupancy_mot_deg = np.zeros((n_rep, len(factor), len(N_deg)))
mean_occupancy_deg_deg = np.zeros((n_rep, len(factor), len(N_deg)))
mean_occupancy_ant_deg = np.zeros((n_rep, len(factor), len(N_deg)))

for i in tqdm.tqdm(range(n_rep)):
    for j, ratio in enumerate(factor):
        for k, n in enumerate(N_deg):
            y0 = [500, 0, 0, 0, 1, n]
            k_array = get_k_array(n, ratio)
            sim_data = simulate_TFsearch(1e1, y0, k_array)
            try:
                motif_bound = np.where(sim_data[:,3]>0)[0]
                first_passage_deg[i,j,k] = sim_data[motif_bound[0], 0]
                mean_occupancy_mot_deg[i,j,k] = compute_mean_occupancy(sim_data, 3, 0)
                mean_occupancy_deg_deg[i,j,k] = compute_mean_occupancy(sim_data, 4, 0)
                mean_occupancy_ant_deg[i,j,k] = compute_mean_occupancy(sim_data, 3, 0, 4)
            except:
                sim_data = simulate_TFsearch(1e4, y0, k_array)
                mean_occupancy_mot_deg[i,j,k] = compute_mean_occupancy(sim_data, 3, 0)
                mean_occupancy_deg_deg[i,j,k] = compute_mean_occupancy(sim_data, 4, 0)
                mean_occupancy_ant_deg[i,j,k] = compute_mean_occupancy(sim_data, 3, 0, 4)
                motif_bound = np.where(sim_data[:,3]>0)[0]
                try:
                    first_passage_deg[i,j,k] = sim_data[motif_bound[0], 0]
                except:
                    continue
                
                    
np.save('simulation_output/ndeg_sensitivity_firstpassage.npy', first_passage_deg)
np.save('simulation_output/ndeg_sensitivity_mean_occupancy_mot.npy', mean_occupancy_mot_deg)
np.save('simulation_output/ndeg_sensitivity_mean_occupancy_deg.npy', mean_occupancy_deg_deg)
np.save('simulation_output/ndeg_sensitivity_mean_occupancy_ant.npy', mean_occupancy_ant_deg)
```
